```python
import pandas as pd 

# Load the data

data = pd.read_csv('clientes_afectados_tiempo_real.csv')

data_historica = pd.read_csv('clientes_afectados_historico_2025.0.csv')

# show quantity of unique or distinct values in the column 'ID_UNICO'

print(data_historica['ID_UNICO'].nunique())
```

```python
#modifica el ID_UNICO con este formato unique_id = f"{fecha_str}-{region}-{comuna}-{empresa}-{afectados}

data_historica['unique_id'] = data_historica['FECHA'].astype(str) + '-' + data_historica['REGION'].astype(str) + '-' + data_historica['COMUNA'].astype(str) + '-' + data_historica['EMPRESA'].astype(str) + '-' + data_historica['CLIENTES_AFECTADOS'].astype(str)
```

```python
# Eliminar las columnas HORA_EXACTA_REPORTE, FECHA_DT, AÃ‘O

data_historica = data_historica.drop(['ID_UNICO'], axis=1)

# renombrar columnas

data_historica = data_historica.rename(columns={'NOMBRE_REGION': 'REGION', 'NOMBRE_COMUNA': 'COMUNA', 'NOMBRE_EMPRESA': 'EMPRESA', 'unique_id': 'ID_UNICO'})
```
